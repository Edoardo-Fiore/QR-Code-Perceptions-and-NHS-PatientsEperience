# QR Code Perceptions and NHS Patients Eperience

## Introduction

Technology has become integral to our daily lives, flooding various fields, including healthcare. In the medical sector, technology holds the potential to revolutionise the way we deliver and manage health services (Liddel et al., 2007). The United Kingdom's National Health Service (NHS) has been proactive in recognising the potential of digitalisation and technology to profoundly enhance its services, intending to reduce costs through streamlined operations, improved efficiency, and, most importantly, enhancing the overall patient experience (Watcher, 2016; Honeyman et al., 2016). Nevertheless, the NHS faces many obstacles in providing its patients with high-quality care (Thorlby et al., 2019). These challenges are influenced by a multitude of factors, including the escalation of healthcare costs due to government budget constraints (ibid), the emigration of medical professionals (Doran at al., 2016; Gauld et al., 2015), and the burden of an ever-increasing patient caseload (Wilkinson, 2023).


## Background and context 

The National Health Service (NHS) encounters various challenges, as adopting new technologies proves demanding even for established organisations (Yang, 2009). Particularly within public sectors like healthcare (Healthcare Industries Task Force, 2004), the NHS, established in 1948, has faced financial issues, mismanagement, and resource cutbacks (Appleby et al., 2014; Gainsbury, 2016). Staff-related challenges, such as pay concerns (Sharma, 2023), impact staff retention (Ratwatte, 2023), and the workforce faces stress-related absences (Ravalier et al., 2020; Sizmur & Raleigh, 2018). Operational inefficiencies and misallocation of budgets indicate a need for transformation and efficient strategies, including technology integration (Appleby et al., 2014).

Patient experience challenges in the NHS include accessibility, data security, and service inequality (Exworthy & Peckham, 2006; Ghafur et al., 2019; Laudicella, 2012). Waiting times, a persistent concern, saw challenges during the COVID-19 pandemic, contributing to a backlog of over 4.4 million patients (Charlesworth et al., 2020). The absence of uniform service standards, effective leadership, and a culture empowering physicians contribute to mismanaged queue times (Godden & Pollock, 2009; Laudicella, 2012; Frost, 1980). Patient confidence and satisfaction levels reached a 25-year low in 2021 (Aiken et al., 2021), emphasising the need for innovative solutions (Johannessen & Alexandersen, 2018).

To address these issues, the NHS is moving towards a user-centred healthcare service, promoting collaboration and integrating users as active team members (Graham & Finn, 2011). Challenging existing organisational culture and administrative processes is crucial (Konteh et al., 2011). Data collection on patient sentiments, facilitated by advancements in digital technology and a growing feedback culture, allows for improved understanding and service enhancement (Ramsey et al., 2019). However, there is a need for systematic strategies to effectively utilise patient experience data (Locock et al., 2020).

## Survey design

The survey, aiming to gather data concerning research questions 1 and 2, adopts a user-centred approach, featuring four sections – Consent, Demographics, Participant & NHS Interactions, and QR Code Interaction Insights. Questions have been meticulously crafted with simplicity and logical wording in mind, avoiding overthinking and confusion for participants (Kasunic, 2005). With a set of 16 questions, the survey ensures a maximum completion time of 5/7 minutes to optimise response rates (Deutskens et al., 2004; Sammut et al., 2021) and leveraging the efficiency of Likert (1932) scales, it captures a total of 45 data points. This streamlined approach maintains a balance between gathering valuable insights and respecting participants' time and engagement levels (Passmore et al., 2002). Within the introduction section (figure 1), the anonymised survey clearly states the research objectives for the participant’s ease (McNeeley, 2012) and transparency. Furthermore, a light blue colour has been picked as the theme of the survey to induce a calm and relaxing feeling for the participants (O’Connor, 2011), making it more visually appealing (Stern et al., 2014) as well as aligning with the recognised colour tone identity of the NSH (Thompson, 2001).

Considerable care has been taken to develop a demographic section that prioritises participant comfort, inclusivity, and non-invasiveness. This involves incorporating a "prefer not to say" option, allowing participants to refrain from responding and including open-ended questions to provide participants with the freedom to express their identities in a manner that feels most fitting. The questions related to ethnicity are framed carefully to avoid subtle discrimination or collateral harm (Nadal et al., 2011), furthermore, terms like "ethnicity" and "race" have been omitted, aligning with a modern and sensitive approach (Cohn, 2015). When inquiring about gender, participants have a range of common options to choose from, with the added flexibility to specify their identity or opt not to respond, in contrast to Moody et al. (2013). The overall design of the demographic section is characterised by inclusive terminology and the option for participants to either withhold information or voluntarily specify it in a manner that best suits them, combining multiple-choice questions with open-ended options (Hughes et al., 2016).

## 2. NHS Perceptions 

We are now going to create visualisations regarding: 

1. Overall NHS perceptions regarding: 
    - Personal Experience 
    - Direction NHS is heading 
    - Brand and Reputation
    - NHS perceived Value 
2. Waiting times: 
    - Appointment waiting times satisfaction level  

```python
#Overalll NHS Perceptions Vis 

# Extracting columns with ordinal values
ordinal_columns = df.loc[(df['gender'] == 'Female') | (df['gender'] == 'Male')]
ordinal_columns = ordinal_columns[['NHS_brand', 'NHS_value', 'patient_experience', 'NHS_direction']]

# Melting columns
ordinal_columns = ordinal_columns.melt(var_name='reasonings', value_name='rating')

# Converting columns to categorical
ordinal_columns['reasonings'] = ordinal_columns['reasonings'].astype('category')
ratings = ['Extremely dissatisfied', 'Somewhat dissatisfied', 'Neutral', 'Somewhat satisfied', 'Extremely satisfied']
ordinal_columns['rating'] = ordinal_columns['rating'].astype(pd.CategoricalDtype(categories=ratings, ordered=True))

# Creating a table for each ordinal value, before ordering
percentage_statements = pd.crosstab(ordinal_columns['reasonings'], ordinal_columns['rating'], normalize='index') * 100
percentage_statements.sort_values(by='Extremely satisfied', inplace=True)

# Plotting the horizontal stacked bar chart
ax = percentage_statements.plot.barh(stacked=True)

# Manually setting the legend outside of the box
handles, labels = ax.get_legend_handles_labels()
ax.legend(reversed(handles), reversed(labels), bbox_to_anchor=(1.0, 1.0))

# Setting chart title and labels
ax.set_title('Percentage of NHS Perceptions based on Ordinal Values')
ax.set_xlabel('Percentage of Responses')

# Display the plot
plt.show()
```

```python
#Waiting Times Vis

# We will aggregate the responses for 'Waiting times' and create a bar chart for this column
waiting_times_column = 'waiting_times'
waiting_times_counts = df[waiting_times_column].value_counts()

# Create a bar chart for the 'Waiting times' responses
plt.figure(figsize=(10, 6))
sns.barplot(x=waiting_times_counts.index, y=waiting_times_counts.values, hue=waiting_times_counts.index)

# Add title and labels
plt.title('Perception of NHS Waiting Times')
plt.xlabel('Level of Satisfaction')
plt.ylabel('Number of Responses')
plt.xticks(rotation=45)  # Rotate labels to avoid overlap

# Show the plot
plt.tight_layout()  # Adjust layout to fit title and x-labels
plt.show()
```

## 2.1 NHS Perception Findings

The survey data on NHS appointment waiting times reveals a spectrum of satisfaction levels among respondents, providing insights into the perceptions of the NHS across multiple dimensions. In terms of brand and reputation, 39% of respondents express extreme satisfaction, while 15% report being extremely dissatisfied. The evaluation of value shows a similar trend, with 37% expressing extreme satisfaction and 13% indicating extreme dissatisfaction. Personal experiences with the NHS follow a comparable pattern, with 35% being extremely satisfied and 15% extremely dissatisfied. Regarding waiting times, 34% are extremely satisfied, contrasting with 14% who are extremely dissatisfied. Finally, in considering the direction the NHS is heading, 38% are extremely satisfied, and 14% are extremely dissatisfied. These findings underscore the diversity of opinions within the surveyed population, indicating a mix of positive and negative sentiments across different aspects of the healthcare system.
Predominantly over 60 percent remain dissatisfied with the current waiting times. This underscores the necessity for clearer delineation in survey questions, specifying the type of service under consideration (phone consultations, GP appointments, specialist hospital appointments). Addressing this clarity issue will enhance the accuracy and reliability of the survey findings.

## 3. QR Code Perception 

We are now going to create visualisations regarding: 

1. Have they used QR code before? 
2. Participants percetion regarding QR codes on: Ease os use, conveniency, QR code for appointment managements, QR code to improve patient experience, QR code to improve appointment management. 
3. QR code to improve patients experience considering participants age groups

```python
#QR code usage in healthcare setting Vis 

plt.figure(figsize=(10, 6))
sns.countplot(x='age_group', hue='QR_used_before', data=df)
plt.title('Distribution of Age Groups by QR Code Usage')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Used QR Code Before')
plt.show()
```

```python
#QR code perception Vis

# Selecting the confirmed relevant columns for the stacked bar chart
confirmed_columns_qr_perception = [
    'QR_ease_of_use',
    'QR_conveniency',
    'QR_appointments',
    'QR_UX',
    'QR_appointments_sentiment']

# Aggregating counts for each response in the confirmed columns
qr_perception_confirmed_aggregated = df[confirmed_columns_qr_perception].apply(pd.Series.value_counts).fillna(0)

# Plotting the stacked bar chart with the confirmed columns
plt.figure(figsize=(12, 8))
qr_perception_confirmed_aggregated.T.plot(kind='bar', stacked=True, ax=plt.gca(), colormap='viridis')

# Adding title and labels
plt.title('Perceptions of QR Codes in Healthcare')
plt.xlabel('QR Code Feature')
plt.ylabel('Number of Responses')
plt.xticks(rotation=45)  # Rotate labels to avoid overlap

# Show the plot
plt.tight_layout()  # Adjust layout to fit title and x-labels
```

```python
#QR Code UX Vis

# Extracting QR code usage
# Count the occurrences of each social media platform in each age group
age_QR_use = df.groupby(['age_group', 'QR_UX']).size().unstack(fill_value=0)

# Plotting the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(age_QR_use, annot=True, cmap='viridis')
plt.title('Heatmap of QR code Usage Across Age Groups')
plt.ylabel('Age Group')
plt.xlabel('Can QR Codes improve patients experience?')
plt.show()
```

The data indicates varying responses to the question of whether QR codes could improve patients' experience, categorised by age groups. Participants aged 18-25 generally agree or strongly agree with the statement. In contrast, respondents aged 36-45 show mixed opinions, with some strongly agreeing and others disagreeing or expressing neutrality. Those aged 46-55 and 56-65 mostly agree or strongly agree, while individuals over 65 tend to be neutral or disagree. The data suggests a diverse range of perspectives on the potential impact of QR codes on enhancing the patient experience, influenced by age-related factors.

## 3.1 QR Code Study FIndings 

The survey data on the usage and perception of QR codes for healthcare-related purposes reveals a range of responses among participants. It's noteworthy that a significant portion of respondents has not utilised QR codes for healthcare, which is intriguing given the widespread use of QR codes during the pandemic, particularly for displaying vaccination status in the UK. Similar to the NHS waiting times section, this might be attributed to the question's broad and unspecific phrasing. Refining the question to inquire specifically about QR code usage within hospitals or for specific purposes could yield more accurate and insightful data. 

For those who have used QR codes, perceptions of user-friendliness vary, with a majority considering QR codes highly user-friendly. Participants generally agree that QR codes are easy to use, convenient, and have the potential to enhance the management of appointments. Notably, some respondents strongly agree with the idea that QR codes can improve the overall patient experience, although a small percentage disagrees with these positive perceptions. Feedback on directions within hospitals is mixed, indicating diverse opinions on clarity and ease of follow-up. In summary, while the overall data suggests a positive inclination toward QR code usage, nuanced views exist regarding their utility and effectiveness, emphasising the importance of specific and targeted survey questions.

However, it's crucial to acknowledge a potential bias in the data due to a predominantly young population, as reflected in the survey. This age skewness could impact the overall user experience (UX) perceptions, as younger individuals may find technology, including QR codes, more familiar and user-friendly. As a result, the positive perceptions might not be representative of the broader population, and accessibility concerns could arise for older demographics or those less accustomed to technology. Being mindful of this potential bias is essential when interpreting the findings, emphasising the need for diverse participant representation in future surveys to ensure a comprehensive understanding of QR code usage in healthcare.

```python
%%js

// Run this cell to update your word count.

function wordcount() {
    let wordCount = 0
    let extraCount = 0
    let mainBody = true

    let cells = Jupyter.notebook.get_cells()
    cells.forEach((cell) => {
        if (cell.cell_type == 'markdown') {
            let text = cell.get_text()
            // Stop counting as main body when get to References or Appendices.
            if (text.startsWith('## References') ||
                text.startsWith('## Appendices')) {
                mainBody = false
            }
            if (text.startsWith('## Word Count')) {
                text = ''
            }
            if (text) {
                let words = text.toLowerCase().match(/\b[a-z\d]+\b/g)
                if (words) {
                    let cellCount = words.length
                    if (mainBody) {
                        wordCount += cellCount
                    } else {
                        extraCount += cellCount
                    }
                }
            }
        }
    })
    return [wordCount, extraCount]
}

let wc = wordcount()
element.append(`Main word count: ${wc[0]} (References and appendices word count: ${wc[1]})`)
```
